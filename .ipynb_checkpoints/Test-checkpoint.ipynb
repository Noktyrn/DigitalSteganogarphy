```python
import os
from Utils.utils import custom_plot
from Utils.StegoSet import StegoSet
from os import listdir
from os.path import isfile, join
from Algorithms.PM1 import insert_message_PM1, extract_message_PM1
from Algorithms.iPM1 import insert_message_iPM1
from Algorithms.AddEmb import dist_func
import numpy as np
from Utils.utils import cut_image_into_blocks, get_dct_coefs, quantize_dct_blocks, qm, get_image_from_dct_coefs, concatenate_image, PSNR, dequantize_dct_blocks
from Algorithms.StegoAlgorithm import StegoAlgorithm as algo
from Utils.Experimenter import Experimenter
import matplotlib.pyplot as plt
```

```python
mypath = os.getcwd() + "/images/grey/"
onlyfiles = [mypath + f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles.remove(mypath + '.DS_Store')

my_set = StegoSet(onlyfiles[:1], init_way='files')
```

```python
def create_emb_map(positions, n=8):
    res = [[0 for _ in range(n)] for _ in range(n)]
    for p in positions:
        res[p[0]][p[1]] = 1
    return res


def embed_message_into_block(point, mes, positions, b):
    I_log = []
    for p in positions:
        I = np.round(b[p[0]][p[1]])
        I_log.append(I)
        if point >= len(mes):
            break
        if I < 0:
            b[p[0]][p[1]] -= 1
        elif I == 0:
            b[p[0]][p[1]] -= mes[point]
            point += 1
        elif I == 1:
            b[p[0]][p[1]] += mes[point]
            point += 1
        else:
            b[p[0]][p[1]] -= mes[point]
            
    return point, b, I_log


def get_message_from_block(mes, positions, b, emb_l):
    for p in positions:
        if emb_l <= 0:
            break
        I = np.round(b[p[0]][p[1]])
        if I == 0 or I == 1:
            emb_l -= 1
            mes.append(0)
        elif I == -1 or I == 2:
            emb_l -= 1
            mes.append(1)
        
        if I >= 2:
            b[p[0]][p[1]] -= 1
        elif I <= -1:
            b[p[0]][p[1]] += 1
    
    return mes, b, emb_l


def get_capacity(positions, block_vector):
    cap = 0
    
    for b in block_vector:
        for p in positions:
            if np.round(b[p[0]][p[1]]) == 0 or np.round(b[p[0]][p[1]]) == 1:
                cap += 1
    
    return cap
    

def insert_message_AddEmb(image, message):
    #step1
    blocks = cut_image_into_blocks(image)
    dct_blocks = quantize_dct_blocks(get_dct_coefs(blocks))
    
    #step2
    b_vector = dct_blocks.reshape(4096, 8, 8)
    
    n = len(b_vector[0])
    num_blocks = len(dct_blocks)

    #step3
    Q = [[] for _ in range(n)]
    for i in range(num_blocks):
        u = i // n
        v = i %  n
        Q[u].append(dist_func(u, v, b_vector))
    
    #step4
    k_coords = []
    for i in range(n):
        for j in range(n):
            k_coords.append([i, j])
    
    k_mins = sorted(k_coords, key=lambda x: Q[x[0]][x[1]])
    
    for k in range(1, len(k_mins)+1):
        embedding_positions = k_mins[:k]
        cap = get_capacity(embedding_positions, b_vector)
        l = len(message)
        emb_l = l
        max_psnr = 0
        best_emb = None
        best_pos = None
        
        if cap >= l:
            extracted_message = []
            
            point = 0
            for idx in range(len(b_vector)):
                start_point = point
                point, temp_b, I_log = embed_message_into_block(point, message, embedding_positions, b_vector[idx])
                b_vector[idx] = temp_b
                extracted_message, temp_b_2, emb_l = get_message_from_block(extracted_message, embedding_positions, b_vector[idx], emb_l)
                if point >= l:
                    break
    
            new_image = concatenate_image(get_image_from_dct_coefs(dequantize_dct_blocks(b_vector.reshape(64, 64, 8, 8)))
            new_psnr = PSNR(new_image, image)
            
            if new_psnr > max_psnr:
                print("Old psne: {} New psnr: {}".format(max_psnr, new_psnr))
                max_psnr = new_psnr
                best_emb = new_image
                best_pos = embedding_positions
                
                
            
    return concatenate_image(best_emb), {'route': best_pos, 'mes_len':len(message)}

def extract_message_AddEmb(im, route, mes_len):
    #step1
    blocks = cut_image_into_blocks(image)
    dct_blocks = quantize_dct_blocks(get_dct_coefs(blocks))
    
    #step2
    b_vector = dct_blocks.reshape(4096, 8, 8)
    
    extracted_message = []
            
    point = 0
    for idx in range(len(b_vector)):
        extracted_message, temp_b_2, mes_len = get_message_from_block(extracted_message, embedding_positions, b_vector[idx], mes_len)
        if emb_len <= 0:
            break
    
    return extracted_image
```

```python
np.set_printoptions(precision=3)
for i in my_set:
    m = [0,1,1,0,1]
    im, params = insert_message_AddEmb(i, m)
    mes = extract_message_AddEmb(im, **params)
    print("Wrote: {} \n Got: {}".format(m, mes))
    break
```
